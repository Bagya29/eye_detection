```python
import cv2
import matplotlib.pyplot as plt
img=cv2.imread(r'C:\DL projects\eye_detection\WhatsApp Image 2023-12-10 at 11.49.23 PM.jpeg')
img
```

```python
plt.imshow(img)
```

```python
cv2.imshow('Bagya',img)
cv2.waitKey(5000)
cv2.destroyAllWindows()
```

```python
conv_img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.imshow(conv_img)
```

```python
eye_cascade=cv2.CascadeClassifier(r'C:\DL projects\eye_detection\haarcascade_eye.xml')
eye_cascade
```

```python
img_gray=cv2.cvtColor(conv_img,cv2.COLOR_RGB2GRAY)
img_gray
```

```python
eye=eye_cascade.detectMultiScale(img_gray,minNeighbors=50)
eye
```

```python
len(eye)
```

```python
for (x,y,w,h) in eye:
    cv2.rectangle(conv_img,(x,y),(x+w,y+h),(0,255,0),4)
```
